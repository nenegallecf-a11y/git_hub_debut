```python
# pour que ces codes fonctionnent il faut bien evidemment executer toutes les ligne de code du notebook de la séance 4 
# #exercice1
full = full.dropna(subset=["ventes_k€", "taux_inflation", "chomage_pourcent"]) #supprimer les ligne dont au moins une des colonnes contient une valeur manquantes

res = [] # je défini une liste vide pour pouvoir y stocker mes coefficient de corrélation (ca me servira également quand je vais creer mon dataframe final)

for reg in full["region"].unique(): # ici j'ai creer une boucle pour calcluler pour CHAQUE REGION les coefficients de correlation demandés 
    sub = full[full["region"] == reg] #filtre le DataFrame pour ne garder que les lignes d’une région spécifique ((reg)
    corr_ventes_inflation = sub["ventes_k€"].corr(sub["taux_inflation"]) # ici je calcule mes coefficient de correlation (coefficient de correlation de pearson)
    corr_ventes_chomage = sub["ventes_k€"].corr(sub["chomage_pourcent"])
    res.append([reg, corr_ventes_inflation, corr_ventes_chomage]) # ici j'ajoute pour la region reg les coefficient que je vient de calculer dans ma liste 
# et la boucle va tourner pour la region suivanteainsi de suite....
# enfin je cré mon tableau final 
corrélations = pd.DataFrame(res, columns=["region", "corr_ventes_inflation", "corr_ventes_chomage"])# ici j'ai creer un tableau qui contient en lignes les resultats de mes calculs de corrélations et en colonnes: (corr_ventes_inflation et corr_ventes_chomage)
corrélations.head()
```
